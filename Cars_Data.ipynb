```python
# Import pandas, matplotlib  which are required for data visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Access Data using read_csv method and store in variable.
df = pd.read_csv("CarsData.csv")
df.head() # Get first 5 rows of data.
```

```python
df.info() # Know all columns datatypes using .info()
```

```python
describe = df.describe() # use statistics function for quick review of data.
print(describe)
```

# Plot 1: Histogram



```python
def plot_histogram(): # Define function for make a histogram
    # Adjust figure size
    plt.figure(figsize=(8,5))

    # Make a histogram plot using matplotlib
    plt.hist(df['fuelType'], color='lightblue')

    # Add title, xlabel, ylabel
    plt.title('Distribution of Car fuelType')
    plt.xlabel('Fuel Type')
    plt.ylabel('Number Of Cars')

    # Show plot
    plt.show()

# call function
plot_histogram()
```

# Plot 2: Scatter Plot

```python
# Define function
def plot_scatter():

    # Adjust figure size
    plt.figure(figsize=(10,6))

    # Make a scatter plot using matplotlib
    plt.scatter(df['mileage'], df['price'], alpha=0.5)

    # Add Title, xlabel, ylabel
    plt.title('Mileage vs Price')
    plt.xlabel('Mileage')
    plt.ylabel('Price')

    # Show grid lines
    plt.grid(True)

    # show plot
    plt.show()

# call function
plot_scatter()
```

# Plot 3: Violin Plot

```python
def plot_violin():
    # Adjust figure size
    plt.figure(figsize=(8, 6))

    # Data preparation
    data = df['price'].dropna().values
    parts = plt.violinplot(data, showmeans=False, showmedians=True)

    # Customizing the violin plot
    for pc in parts['bodies']:
        pc.set_facecolor('lightblue')
        pc.set_edgecolor('black')
        pc.set_alpha(0.7)

    # Add title, ylabel
    plt.title('Violin Plot of Car Prices')
    plt.ylabel('Price')
    plt.grid(True)
    plt.show()

plot_violin()
```

```python
# Step 1: Convert all columns to numeric, coercing errors
df = df.apply(pd.to_numeric, errors='coerce')

# Step 2: Drop rows with NaN values to ensure clean data for correlation
df = df.dropna()

# Step 3: Calculate the correlation matrix
correlation_matrix = df.corr()

# Step 4: Print the correlation matrix
print(correlation_matrix)
```
